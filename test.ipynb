```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def integrand(x, T, Delta):
    return 1 / (np.cosh(np.sqrt(x**2 + Delta**2) / (2 * T))**2)

def Y_T(T, Delta):
    integral, _ = quad(integrand, 0, np.inf, args=(T, Delta))
    return integral / T


plt.figure(figsize=(8, 5))
# Define parameters
T_values = np.linspace(0.01, 5, 1000)  # Temperature values

for Delta in [0,0.1,0.5,1]:
    # Delta = 1.0  # Fixed parameter

    # Compute Y(T) for different T
    Y_values = [Y_T(T, Delta) for T in T_values]

    # Plot the results

    plt.plot(T_values, Y_values, label=f"Delta = {Delta}")


plt.xlabel("T")
plt.ylabel("Y(T)")
plt.title("Plot of Y(T) for different T")
plt.legend()
plt.grid()
plt.show()
```
