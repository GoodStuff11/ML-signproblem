```julia
using LinearAlgebra


# The number of components in a_i and M_i

DIM = 2
N = 1000 # Order of truncation
# Make up some coeffcients
v_1, v_2 = ones(4), ones(4)
M = [0.01 * rand(4, 4) for _ in 1:DIM]
# The objective function.
objective = (a) -> v_1' * exp(sum(i->a[i] * M[i], 1:DIM)) * v_2


# The analytic derivative, truncated to N = 1000.

# This is a slow implementation prioritizing clarity.

function d_objective(a)

    X = sum(i->a[i] * M[i], 1:DIM) # The argument to exp.

    1/N * map(j-> sum(i-> v_2' * (I + X/N)^(N - i) * M[j] * (I + X/N)^i * v_1, 1:N), 1:DIM) 

end


# Compare with finite differece.

finite_difference_check = [ 

    objective([1, 1]) - objective([0.999, 1]),

    objective([1, 1]) - objective([1, 0.999]) 

] * 1000


isapprox(d_objective([1, 1]), finite_difference_check, atol=1e-3)
```
