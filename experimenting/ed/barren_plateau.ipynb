```julia
function extract_info(s::AbstractString)
    # 1. Split the start string and the bracketed content.
    m = match(r"^(.+?)\s+\((.*)\)$", s)
    if m === nothing
        return nothing
    end
    
    start_str = m.captures[1]
    bracket_content = m.captures[2]
    
    # 2. Check for `std=` and `exp`
    m_std = match(r"std=([0-9.]+)", bracket_content)
    m_exp = match(r"([0-9]+)\s*exp", bracket_content)
    
    std_val = m_std !== nothing ? parse(Float64, m_std.captures[1]) : nothing
    exp_val = m_exp !== nothing ? parse(Int, m_exp.captures[1]) : nothing
    
    # 3. Determine the label
    local label::String
    if m_std !== nothing
        # If std= exists, extract string before (std=...)
        m_pre_std = match(r"^(.*?)\s*\(std=", bracket_content)
        label = m_pre_std !== nothing ? String(strip(m_pre_std.captures[1])) : ""
    else
        # Otherwise, remove any "(N exp)" and surrounding whitespace
        cleaned = replace(bracket_content, r"\s*\([0-9]+\s*exp\)" => "")
        label = String(strip(cleaned))
    end
    
    return (start=start_str, std=std_val, exp=exp_val, label=label)
end
for col_name in names(df)
    println(extract_info(col_name))
end
```

```julia
using Plots

# Create the main plot (no legend)
p_main = plot(xlabel="System size", 
              ylabel=L"\frac{1}{L}||\nabla\langle H\rangle||^2", 
              legend=false, yscale=:log10,
              left_margin=5Plots.mm,
              right_margin=-15Plots.mm,
              bottom_margin=5Plots.mm, yticks=10 .^(-5.:0.))

# Create a blank subplot for the legend
p_legend = plot(framestyle=:none, legend=:left,
            right_margin=40Plots.mm,
            bottom_margin=5Plots.mm)
x_vals = sort(unique(df[!, "n modes"]))

c_i = 1
cmap = palette(:seaborn_flare_gradient, 3)
for col_name in names(df)
    info = extract_info(col_name)
    if info === nothing
        continue
    end

    num_exp = info[:exp]
    std = info[:std]
    label = info[:label]
    metric = info[:start]

    if !(std == 0.1 || num_exp === nothing) # || metric == "E[norm(g)^2]")
        continue
    end
    
    if !(metric in ["E[norm(g)^2]", "norm(g)^2"])
        continue
    end
    y = []
    for D in x_vals
        sub_df = filter(row -> row["n modes"] == D, df)
        m = sub_df[1, col_name]/sub_df[1, "Param Count"]
        push!(y, m)
    end
    if num_exp == nothing
        c = :black
        linestyle = :dash
    else
        c = cmap[c_i]
        linestyle = :solid
        label = "Slater ref"
    end
    # Plot the actual data on the main plot (label=false to keep it clean)
    plot!(p_main, x_vals./2, y, label=false, c=c, linestyle=linestyle)
    
    # Plot an empty series on the legend canvas (keeps colors synced automatically)
    plot!(p_legend, Float64[], Float64[], label="$metric num exp=$num_exp std=$std label=$label", c=c, linestyle=linestyle)
    c_i += 1
end

# Combine them: 70% width for the main plot, 30% for the legend area
# Increased overall width to 900px to ensure the long legend labels don't get cut off
p_final = plot(p_main, p_legend, layout=@layout([a{0.7w} b]), size=(900, 500))

display(p_final)
```

```julia
using Statistics

# Unique system dimensions
x_vals = sort(unique(df[!, "n modes"]))

# Define the cases: (label, column_name, is_gradient_to_be_squared, horizontal_jitter_factor)
cases = [
    (L"\frac{1}{L}\mathbb{E}[||\nabla \langle H\rangle||^2]"*" random initalization", "E[norm(g)^2] (ϵ=0s basis, ϵ=10.0)","Std[norm(g)^2] (ϵ=0s basis, ϵ=10.0)", -0.1),
    (L"\frac{1}{L}||\nabla \langle H\rangle||^2"*" single random sample", "norm(g)^2 (random basis, ϵ=0.0)", nothing, 0.0),
    (L"\frac{1}{L}||\nabla\langle H\rangle||^2"*" nn initialized", "norm(g)^2 (nn gs, ϵ=0.0)",nothing, 0.1)
]

p = plot(yscale=:log10,
         xlabel="# of modes", 
         ylabel=L"\frac{1}{L}||\nabla\langle H\rangle||^2", 
         legend=:bottomleft)

for (label, col, std_col, jitter) in cases
    means = Float64[]
    stds = Float64[]
    
    for D in x_vals
        sub_df = filter(row -> row["n modes"] == D, df)
        m = sub_df[1, col]/sub_df[1, "Param Count (K)"]
        s = 0.0
        if !isnothing(std_col)
            s = sub_df[1, std_col]/sub_df[1, "Param Count (K)"]
        end
        
        push!(means, m)
        push!(stds, s)
    end
    
    # Asymmetric error bars to prevent negative values on the log scale
    lower_err = means .- max.(10^(-4), means .- stds)
    yerr = (lower_err, stds)
    
    scatter!(p, x_vals .+ jitter, means, yerror=yerr, label=label, markersize=6)
end

display(p)
```

```julia
using LsqFit

# Extract x and y values
x = df[!, "Hilbert Dim (D)"]
y_nn = 1 ./ df[!, "Max Var (NN, ϵ=0.005)"]
# "Max Var (Rand, ϵ=0.1)"   "Max Var (Rand, ϵ=10.0)"]
y_rand = 1 ./ df[!, "Max Var (0s+RandRef, ϵ=0.005)"]

x_range = [minimum(x), maximum(x)]
# Define the linear model function where p[1] is intercept and p[2] is slope
@. model(x, p) = p[1] + p[2] * x
# Initial guess for parameters [intercept, slope]
p0 = [0.0, 0.0]
# Fit the curve
fit_nn = curve_fit(model, x, y_nn, p0)
fit_rand = curve_fit(model, x, y_rand, p0)

# Retrieve coefficients (beta) and standard errors
beta_nn = coef(fit_nn)
beta_rand = coef(fit_rand)
se_nn = stderror(fit_nn)
se_rand = stderror(fit_rand)



println("nn: ($(beta_nn[2])±$(se_nn[2]))x + ($(beta_nn[1])±$(se_nn[1]))")
println("rand: ($(beta_rand[2])±$(se_rand[2]))x + ($(beta_rand[1])±$(se_rand[1]))")



# 1. Plot the scatter points
scatter(x, y_nn, label="Neural network", ylabel=L"1/\textrm{max}_i\textrm{Var}[\partial_i\langle H\rangle]", xlabel="Hilbert space dimension")
scatter!(x, y_rand, label="Random")

# 2. Plot the fitted lines
plot!(x_range, model(x_range, beta_nn), label="Neural network fit", linewidth=2)
plot!(x_range, model(x_range, beta_rand), label="Random fit", linewidth=2)
```

```julia
using CSV
using DataFrames
using Plots
using LaTeXStrings
using JLD2
using Lattices
using Markdown

include("extract_losses.jl")

using SparseArrays
include("utility_functions.jl")
include("ed_objects.jl")
include("ed_functions.jl")


using DataFrames
using JLD2
using Plots



folder1 = "nn_test_data"
folder2 = "nn_test_data_random"

subfolders = ["N=(4, 5)_3x3_2", "N=(4, 4)_3x3", ]
nn_labels = [
    "2x2_only_one_minus_loss_power_10",
    "pure_mild_mid_narrow",
]
u_indices = [25,39,49,53]
cmap = palette(:managua, length(u_indices))

dfs = []

# Load interaction strengths
U_values = load(joinpath("nn_test_data", "N=(4, 5)_3x3_2", "meta_data_and_E.jld2"))["dict"]["meta_data"]["U_values"]

# for subfolder in subfolders
#     push!(dfs, collect_scaling_results(
#         "logs"; subfolder=subfolder, 
#         u_values=U_values[u_indices],
#         nn_labels=nn_labels,
#     ))
# end
# df2 = vcat(dfs...)
# desired_columns = ["Baseline loss", "Pred loss", "randx10 min loss", "mean loss","max loss"]
# df2 = df2[:, ["nn_name","folder", "U-va", desired_columns...]]
# df2[!,"folder"] = [subfolders[findfirst(==(item), subfolders_foldername)] for item in df2[!, "folder"]]

# println(dataframe_to_markdown(df2))


df = DataFrame(
    u_idx = Int[],
    U_val = Float64[],
    folder = String[]
)
df[!, "initial/optimized"] = String[]
for label in nn_labels
    df[!, label] = Float64[]
end
df[!, :random] = Float64[]

loss_labels = ["energy", "overlap"]
for subfolder = subfolders
    metadata = load(joinpath(folder1, subfolder, "meta_data_and_E.jld2"))["dict"]
    k_min_large = find_best_energy_sector(metadata["E"], U_values; verbose=false)
    # 1. Initialize empty DataFrame with the correct schema

    p = [
        plot(xlabel="Iterations", ylabel=L"E-E_g",  title=subfolder, legend=nothing, yscale=:log10),
        plot(xlabel="Iterations", ylabel="Overlap loss",  title=subfolder, yscale=:log10, legend=nothing)
    ]

    for (k, u_idx) in enumerate(u_indices)
        U_val = round(U_values[u_idx]; digits=5)
        # hline!(p[1], [metadata["E"][k_min_large][u_idx]], c=cmap[k], linestyle=:dot, label="U=$U_val true energy")
        for (p_i,(loss_metric, loss_metric_label)) in enumerate(zip(["loss_energy_", ""], loss_labels))
            loss_to_overlap(x, loss_metric_label) = loss_metric_label == "overlap" ? 1 - x : x

            
            # Temporary vectors to store losses for the NN labels
            nn_initials = Float64[]
            nn_opts = Float64[]
            
            # 2. Extract and plot losses for the neural network models
            for (label_i, label) in enumerate(nn_labels)
                linestyles = [:solid, :dashdot]
                file = "unitary_map_energy_symmetry=false_$(subfolder[1:8])_$(loss_metric)nn_$(label)_u_$(u_idx).jld2"
                all_losses, was_multistart, opposite_val = extract_losses_from_file(joinpath(folder1, subfolder, file); order_idx=1)
                println("optimized $(loss_labels[p_i]) NN U=$U_val: $(loss_labels[p_i])=$(loss_to_overlap(all_losses[end], loss_metric_label)) $(loss_labels[3-p_i])=$(opposite_val)")
                println("initial $(loss_labels[p_i]) NN U=$U_val: $(loss_labels[p_i])=$(loss_to_overlap(all_losses[2], loss_metric_label))")
                println()
                push!(nn_initials, all_losses[1])
                push!(nn_opts, all_losses[end])
                if loss_metric_label == "energy"
                    all_losses[2:end] = all_losses[2:end] .- metadata["E"][k_min_large][u_idx]
                end
                plot!(p[p_i], all_losses[2:end], label="U=$U_val $label", linewidth=2, linestyle=linestyles[label_i], c=cmap[k])
                scatter!(p[p_i], [0, length(all_losses)],[all_losses[2], all_losses[end]], marker=:xcross, c=cmap[k], label=nothing)
            end
            
            # 3. Extract and plot losses for the random model
            file = "unitary_map_energy_symmetry=false_$(subfolder[1:8])_$(loss_metric)random_multistart_u_$(u_idx).jld2"
            all_losses, was_multistart, opposite_val = extract_losses_from_file(joinpath(folder2, subfolder, file); order_idx=1)
            println("optimized $(loss_labels[p_i]) random U=$U_val: $(loss_labels[p_i])=$(loss_to_overlap(all_losses[end], loss_metric_label)) $(loss_labels[3-p_i])=$(opposite_val)")
            println("initial $(loss_labels[p_i]) random U=$U_val: $(loss_labels[p_i])=$(loss_to_overlap(all_losses[2], loss_metric_label))")
            println()
            rand_initial = all_losses[1]
            rand_opt = all_losses[end]
            
            if loss_metric_label == "energy"
                all_losses[2:end] = all_losses[2:end] .- metadata["E"][k_min_large][u_idx]
            end
            plot!(p[p_i], all_losses[2:end], label="U=$U_val random", linewidth=2, linestyle=:dash, c=cmap[k])
            scatter!(p[p_i], [0, length(all_losses)],[all_losses[2], all_losses[end]], marker=:cross, c=cmap[k], label=nothing)
            # 4. Append row for the "initial" values
            push!(df, (u_idx, U_val, subfolder, "initial", nn_initials..., rand_initial))
            
            # 5. Append row for the "optimized" values
            push!(df, (u_idx, U_val, subfolder, "optimized", nn_opts..., rand_opt))
        end
    end

    # Display the resulting DataFrame
    display(p[1])
    display(p[2])
end
```
